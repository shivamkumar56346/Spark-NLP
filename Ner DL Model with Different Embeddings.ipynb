```
bert_annotator=BertEmbeddings.pretrained().setInputCols(["sentence","token"]).setOutputCol("bert").setCaseSensitive(False)
```

```
test_data=bert_annotator.transform(test_data)
test_data.show(3)
```

```
test_data.select('bert.result','bert.embeddings','label.result').show(5)
```

```
import numpy as np
emb_vector=np.array(test_data.select('bert.embeddings').take(1))
emb_vector
```

```
nerTagger=NerDLApproach().setInputCols(['sentence','token','bert']).setLabelColumn('label').setOutputCol('ner').setMaxEpochs(1) \
.setValidationSplit(0.2).setEvaluationLogExtended(True).setEnableOutputLogs(True).setIncludeConfidence(True)
```

```
pipeline=Pipeline(stages=[bert_annotator,nerTagger])
```

```
ner_model=pipeline.fit(training_data.limit(1000))
```

```
ner_model.stages
```

```
predictions=ner_model.transform(test_data)
predictions.show(5)
```

```
glove=WordEmbeddingsModel().pretrained().setInputCols(["sentence",'token']).setOutputCol("glove").setCaseSensitive(False)
```

```
test_data=glove.transform(test_data.limit(1000))
```

```
test_data.show(5)
```

```
nerTagger=NerDLApproach().setInputCols(['sentence','token','glove']).setLabelColumn('label').setOutputCol('ner').setMaxEpochs(1) \
.setValidationSplit(0.2).setEvaluationLogExtended(True).setEnableOutputLogs(True).setIncludeConfidence(True)
```

```
glove_pipeline=Pipeline(stages=[glove,nerTagger])
```

```
new_ner_model=glove_pipeline.fit(training_data.limit(1000))
```

```
new_ner_model.stages
```

```
small_predictions=new_ner_model.transform(test_data.limit(10))
small_predictions.show(5)
```

```
small_predictions.select('token.result','label.result','ner.result').show()
```
